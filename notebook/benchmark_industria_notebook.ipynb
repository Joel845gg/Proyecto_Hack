```python
# Verificar nulos críticos
critical_cols = ['empresa', 'nombre_plan', 'velocidad_download_mbps', 'precio_plan']
null_counts = df[critical_cols].isnull().sum()

print("Validación de datos críticos:")
for col in critical_cols:
    null_pct = (null_counts[col] / len(df)) * 100
    status = "✅" if null_pct < 5 else "⚠️"
    print(f"  {status} {col}: {null_pct:.1f}% nulos")
```

```python
stats = df.groupby('empresa').agg({
    'nombre_plan': 'count',
    'velocidad_download_mbps': 'mean',
    'precio_plan': 'mean'
}).round(2)

stats.columns = ['cantidad_planes', 'velocidad_promedio', 'precio_promedio']
stats
```
